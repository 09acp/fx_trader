```python
import pandas as pd 
import numpy as np 
import sqlite3 as sql
import datetime
import matplotlib.pyplot as plt

conn = sql.connect('FX_data/database.db')
cur = conn.cursor()
df = pd.read_sql_query("SELECT * FROM profitable_subgroups WHERE Date >= '2009-01-01'",conn) #eda_data_LR
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date'],inplace=True)
conn.close()

df = df[['Target','Prediction','Correct','% Sell','% Buy','Price']] 
df.columns = ['Labels','Pred_Model','Correct_Model','% Sell','% Buy','Close']
df.head(1)
```

#### Trader

```python
def classification_trader (balance, investment_size, prices,\
                        target,prediction,path,\
                        model_name,commission, dates ):
    """
    - Buy and Sell Trader agent 
    - For a single stock

    inputs: starting balance,position_size, df_labels, df_pred
    returns: print info and returns balance_tracker for plots 
    """
    trade_log = pd.DataFrame()
    # LOGS TO KEEP TRACK OF TRADER PERFORMANCE
    log = []; log_returns = []; log_profit = []; a =[]; ac=[]
    for today in range(prices.index.size):
        #break before end Sequence
        if today == len(prices.index)-1: break
        tomorrow = today + 1 
        # CLOSING PRICE
        today_close = prices[today]
        tomorrow_close =  prices[tomorrow]
        # PREDICTED ACTION FOR NEXT DAY
        d ={0:'NEUTRAL',1:'BUY',2:'SELL'}
        action = int(prediction[tomorrow])        # PREDICTIONS
        correct_action = int(target[tomorrow])    # TARGETS
        # DETERMINES HOW MANY SHARES CAN BE BOUGHT GIVEN INVESTMENT_SIZE
        position = investment_size / today_close

        print ('BOUGHT %s SHARES @ %s'%(np.round(position), np.round(today_close)))        
        """
        - Predict what position has to be taken 
        - See whether position has paid-off    
        """
        # BUY AGENT: PROFITS FROM APPRECIATION OF PRICE
        if action == 1: 
            ''' # makes buy only
            profit = 0
            achieved_returns = 0
            log.append(balance); log_profit.append(profit) ; log_returns.append(achieved_returns)
            '''
            achieved_returns = (tomorrow_close - today_close) # HOW MUCH SHARE PRICE ROSE
            profit = (achieved_returns)   * position # SHARE NUMBER * APPRECIATION
            transaction_cost = abs(profit)*commission # TRANSACTION COST
            profit = profit - transaction_cost
            balance += profit
            print('BOUGHT: profit: $%s, return: %s'%(int(profit),np.round(achieved_returns,3)))
            print('Commission: ',np.round(transaction_cost))
            if profit < 0: print( 'MISS__Should have been: %s'%(d[correct_action]))
            log.append(balance); log_profit.append(profit) ; log_returns.append(achieved_returns)
            a.append(action) ; ac.append(correct_action)
            print('CurrentBalance: ', int(balance), '\n') 
        # SELL AGENT: PROFITS FROM DEPRECIATION OF PRICE
        elif action == 0: 
            ''' # makes sell only
            profit = 0
            achieved_returns = 0
            log.append(balance); log_profit.append(profit) ; log_returns.append(achieved_returns)
            '''
            achieved_returns = (today_close - tomorrow_close)
            profit = (achieved_returns)    * position
            transaction_cost = abs(profit)*commission # TRANSACTION COST
            profit = profit - transaction_cost
            balance += profit
            print('SOLD: profit: $%s, return: %s'%(int(profit),np.round(achieved_returns,3)))
            print('Commission: ',np.round(transaction_cost))
            if profit < 0: print( 'MISS__Should have been: %s'%(d[correct_action]))
            log.append(balance); log_profit.append(profit) ; log_returns.append(achieved_returns)
            a.append(action) ; ac.append(correct_action)
            print('CurrentBalance: ',int(balance),'\n')                         
        elif action == 100:
            achieved_returns = 0
            profit = 0
            transaction_cost = 0
            balance += profit
            print('PASSED')
            log.append(balance); log_profit.append(profit) ; log_returns.append(achieved_returns)
            a.append(action) ; ac.append(correct_action)            
            
    trade_log['Date'] = dates
    trade_log['Balance'] = pd.Series(log)
    trade_log['Profit'] = pd.Series(log_profit)
    #trade_log['Returns'] = pd.Series(log_returns)
    trade_log['Action'] = pd.Series(a)
    trade_log['Target'] = pd.Series(ac)
    
    trade_log.set_index('Date',inplace=True)
    # PROFIT PER DAY
    fig = plt.figure(figsize = (9,6))
    trade_log['Profit'].cumsum().plot() 
    plt.title('%s Profit'%(model_name))
    plt.savefig(path+'%s__Profit.png'%(model_name))
    plt.show()
#     # ACHIEVED RETURNS 
#     fig = plt.figure(figsize = (9,6))
#     trade_log['Returns'].plot() 
#     plt.title('%s Returns'%(model_name))
#     plt.savefig(path+'%s__Returns.png'%(model_name))
#     plt.show()
    # BALANCE
#     fig = plt.figure(figsize = (9,6))
#     trade_log['Balance'].plot()
#     plt.title('%s Balance'%(model_name))
#     plt.savefig(path+'%s__Balance.png'%(model_name))
#     plt.show()
    return trade_log

def vizualize_trader_class (dates,
                            prediction,
                            price,
                            df_trader,
                            model_name,
                            path):
#     df.reset_index()['Date'], df['Prediction'], \
#                        df['Price'], df_trading, \
#                        model_name, path

    df_actions = pd.DataFrame()
    df_actions['Date'] = dates
    df_actions.set_index(['Date'],inplace=True)
    # IF PROBABILITIES PRESENT IN AXIS - DISCARD

    df_actions['Action'] = prediction
    #df_actions.reset_index(drop=True,inplace=True)

    buy_signals = prediction.reset_index().index[prediction == 1].to_list()
    sell_signals = prediction.reset_index().index[prediction == 0].to_list()
#     end_balance = int(df_trader.Balance.dropna()[-1])
    profit = np.round(df_trader.Profit.sum(),2)


    fig = plt.figure(figsize = (18,10))
    ' INDEX'
    plt.plot(price, color='black', lw=2.)
    plt.plot(price, '^', markersize=8, color='g', label = 'BUY signal', markevery = buy_signals)
    plt.plot(price, 'v', markersize=8, color='r', label = 'SELL signal', markevery = sell_signals)
    plt.legend()
    plt.xlabel ('Backtest Period')
    plt.ylabel ('Exchange Price USD/EUR')
    plt.title('Trader: %s.  Profit $%s'%(model_name, profit))
    plt.savefig(path+'%s__Trades.png'%(model_name))
    plt.show()
    
    
def summary (df,name_correct): 
    """
    Financial performance metrics for trades
    - Input: df
    - Output: Table of statistics
    """
    report = {
            'Total Profit ($)': df['Profit'].sum(),
            'Mean Profit ($)': df['Profit'].mean(),
            'Volatility (std $)': df['Profit'].std(),
            'Max Loss': df['Profit'].min(),
            'Max Win': df['Profit'].max(),
            'Skewness': df['Profit'].skew(),
            'Kurtosis': df['Profit'].kurt(),
            'Tail Ratio': df['Profit'].quantile(.95) / df['Profit'].quantile(.05),
            'Hit Ratio (%)': df[name_correct].sum()/len(df[name_correct]),
            'Average Win to Average Loss': df['Profit'].loc[df['Profit']>0].mean() / df['Profit'].loc[df['Profit']<0].mean(),
            'Average Profitability Per Correct Trade ($)':  df['Profit'].loc[df[name_correct]==1].mean()
        }
    table = pd.Series(
        report,
        dtype=object
    )
    return table
```

## Data Prep
- Create 3 dataframes according to each of the strategy

```python
conn = sql.connect('FX_data/database.db')
cur = conn.cursor()
df_results = pd.read_sql_query("SELECT * FROM sub_trade_eval",conn)
df_results['Date'] = pd.to_datetime(df_results['Date'])
df_results.set_index(['Date'],inplace=True)
df_results['Pred_Analyst'] = df_results['Subjective_Decision'].astype('int')
conn.close()

df_results['Pred_Match'] = np.where((df_results['Prediction'] == df_results['Pred_Analyst']), 1,0)#Trade:1,Pass:0
df_results.sample(3)
```

```python
df['Pred_Match'] = df_results['Pred_Match']
df['Correct_Match'] = np.where((df['Pred_Match'] == df['Labels']), 1, 0) # Correct: 1, Wrong: 0
df['Pred_Analyst'] = df_results['Pred_Analyst']
df['Correct_Analyst'] = np.where((df['Pred_Analyst'] == df['Labels']), 1, 0) # Correct: 1, Wrong: 0

# Clear sheet blank df to append values 
df_base = df[['Labels','% Sell','% Buy','Close']]

# Create 3 dfs based on strategy
# 1   70 trades 
df_model = pd.concat([df_base,df[['Pred_Model','Correct_Model']]],axis=1)   
"""
With match, only a portion of trades are executed, 
so added 100 to Analyst Prediction to indicate to trading agent
which time periods t to ignore.

Hence temporal relation of prices not affected and profit can be calculated
"""
df_match = pd.DataFrame()
df_match = pd.concat([df_base,df[['Pred_Match','Pred_Analyst']]],axis=1)   
# whether predictions correct 
df_match['Correct_Match'] = np.where((df_match['Pred_Match'] == df_match['Labels']), 1, 0) # Correct: 1, Wrong: 0
# replaces predictions where forecast doesnt match with 100
df_match['Pred_Analyst'].loc[df_match['Pred_Match']!=1] = df_match.loc[df_match['Pred_Match']!=1]['Pred_Analyst'].replace([0,1],100)

# 3   70 trades
df_analyst= pd.concat([df_base,df[['Pred_Analyst','Correct_Analyst']]],axis=1)   
```

## Trading experiments 
- Trade predictions of each of the 3 trading strategies 
- Log profit 

```python
# SET PARAMETERS FOR TRADER
balance = 10000
investment_size = 5000 # TRADES DO NOT EXCEED A FEW HUNDREDS $  -
commission = 0.03 # % COMMISSION PER TRADE
path = 'images/h3_trading/'#'images/trading_log'

# PROFIT FOR MODEL BASED PREDICTIONS 
model_name = 'trader_model'
trader_model = classification_trader(balance, 
                                     investment_size, 
                                     df_model['Close'],   # stock price 
                                     df_model['Labels'],   # answers 
                                     df_model['Pred_Model'],  # predictions
                                     path,
                                     model_name,
                                     commission, 
                                     df_model.reset_index()['Date'] )

# PROFIT FOR MATCHING BASED TRADES
model_name = 'trader_match'
trader_match = classification_trader(balance, 
                                     investment_size, 
                                     df_match['Close'], 
                                     df_match['Labels'],
                                     df_match['Pred_Analyst'], 
                                     path,
                                     model_name,
                                     commission, 
                                     df_match.reset_index()['Date'] )

# PROFIT FOR ANALYST BASED TRADES
model_name = 'trader_analyst'
trader_analyst = classification_trader(balance, 
                                     investment_size, 
                                     df_analyst['Close'], 
                                     df_analyst['Labels'],
                                     df_analyst['Pred_Analyst'], 
                                     path,
                                     model_name,
                                     commission, 
                                     df_analyst.reset_index()['Date'] )
```

```python
# append profit to DF
df_model['Profit'] = trader_model['Profit']
df_match['Profit'] = trader_match['Profit']
df_analyst['Profit'] = trader_analyst['Profit']
```

```python
model_name = 'trader_model'
vizualize_trader_class(df_model.reset_index()['Date'],
                       df_model['Pred_Model'],
                       df_model['Close'],
                       df_model,
                       model_name,
                       path)
summary(df_model,'Correct_Model')
```

```python
model_name = 'trader_match'
vizualize_trader_class(df_match.reset_index()['Date'],
                       df_match['Pred_Analyst'], # either model pred or analyst pred, all forecasts are the same anyway
                       df_match['Close'],
                       df_match,
                       model_name,
                       path)
"""
-The data for the match trader stats is different because only 39 trades where executed from the 70.
-In order calculate truthful averages had to pass only columns with values
"""
summary(trader_match.loc[trader_match['Action']!=100].dropna(),'Target')
```

##### Analyst

```python
model_name = 'trader_analyst'
vizualize_trader_class(df_analyst.reset_index()['Date'],
                       df_analyst['Pred_Analyst'],
                       df_analyst['Close'],
                       df_analyst,
                       model_name,
                       path)
summary(df_analyst,'Correct_Analyst')
```
